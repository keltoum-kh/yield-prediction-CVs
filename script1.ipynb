```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point, Polygon
from sklearn.model_selection import GroupKFold, cross_val_predict
from sklearn import metrics
import xgboost as xgb


workbook = 'field_point.xlsx'
df=pd.read_excel(workbook,'all')
X = df.iloc[:, 5:-1].values  
Y = df.iloc[:, -1].values 


def process_iteration(cell_size):
    geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
    gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")  # CRS for WGS84

    SA = "SA.shp"
    gdf.to_file(SA)
    bounds = gdf.total_bounds
    minx, miny, maxx, maxy = bounds
    rows = int((maxy - miny) // cell_size + 3)
    cols = int((maxx - minx) // cell_size + 3)
    polygons = []
    for i in range(rows):
        for j in range(cols):
            xmin = minx - 1.5 * cell_size + j * cell_size
            xmax = xmin + cell_size
            ymin = miny - 1.5 * cell_size + i * cell_size
            ymax = ymin + cell_size
            polygon = Polygon([(xmin, ymin), (xmin, ymax), (xmax, ymax), (xmax, ymin)])
            polygons.append(polygon)
            

    grid_gdf = gpd.GeoDataFrame({'geometry': polygons}, crs="EPSG:4326")

    grid_shapefile = f"grid_{cell_size}.shp"
    grid_gdf.to_file(grid_shapefile)

    SA = gpd.read_file("SA.shp")

    grid = gpd.read_file(grid_shapefile)

    grid_df = pd.DataFrame(columns=[f'grid_{cell_size}'])

    for index, point in SA.iterrows():
        containing_grid = grid[grid.contains(point.geometry)]

        
        if not containing_grid.empty:
            field_value = containing_grid.iloc[0]['FID']
            grid_df.loc[index] = [field_value]
        else:

            nearest_grid = grid.loc[grid.distance(point.geometry).idxmin()]
  
            field_value = nearest_grid['FID']  
            grid_df.loc[index] = [field_value]


    model = xgb.XGBRegressor()
    block = grid_df[f'grid_{cell_size}'].values
    group_kfold = GroupKFold(n_splits=3)
    block_kfold = group_kfold.split(X, Y, block)

    train_indices, test_indices = [list(traintest) for traintest in zip(*block_kfold)]
    block_cv = [*zip(train_indices, test_indices)]
    Y_xgb = cross_val_predict(model, X, Y, cv=block_cv)
    rscore = metrics.r2_score(Y, Y_xgb)
    mae = int(metrics.mean_absolute_error(Y, Y_xgb))
    rmse = int(np.sqrt(metrics.mean_squared_error(Y, Y_xgb)))
    return rscore, mae, rmse


cell_sizes = np.arange(1, 251) * 0.005
num_iterations = len(cell_sizes)
evaluation_results = []

for i, cell_size in enumerate(cell_sizes):
    print(f"Iteration {i+1}/{num_iterations}: Cell Size {cell_size}")
    rscore, mae, rmse = process_iteration(cell_size)
    evaluation_results.append((rscore, mae, rmse))


evaluation_table = pd.DataFrame(evaluation_results, columns=['R²', 'MAE', 'RMSE'])


```
